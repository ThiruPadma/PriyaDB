```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
file = 'blackmeasles (1).jpg'
image = cv2.imread(file, 0)
#image = cv2.resize(image, (500, 500))
cv2.imshow('leaf1', image)
#plt.imshow(image)
cv2.waitKey(0)
```

```python
def pre_process(image):
    resized = cv2.resize(image, (500,500))
    blurred = cv2.medianBlur(resized, 3)
    image = cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    equalized = cv2.equalizeHist(gray)
    return(equalized)
    
    
    
```

```python
img = pre_process(image)
cv2.imshow("pre", img)
cv2.waitKey(0)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
gmask = cv2.inRange(hsv, (36, 25, 25), (70, 255,255))
igmask = gmask>0
green = np.zeros_like(img, np.uint8)
green[igmask] = img[igmask]
cv2.imwrite("green.jpg", green)
cv2.imshow("Gr", green)
cv2.waitKey(0)
```

```python
#img = cv2.imread(file, 0)
img = image
#img = cv2.resize(img, (800,800))
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
```

```python
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
